```python
#user login check
print("Task 1 - User Login Check")
username = "admin"
password = "1234"
input_username = username
input_password = password
if input_username == username and input_password == password:
  print ("Login Successful")
else:
  print("Invalid Credentials")
print("\n")
```

```python
#pass/fail analyser
print("Task 2 - Pass/ fail Analyser")
marks = [45,78, 90, 33, 60]
pass_count = 0
fail_count = 0
for mark in marks:
  if mark >= 50:
    pass_count += 1
  else:
    fail_count += 1
print("total students: ", len(marks))
print("number of students passed: ", pass_count)
print("number of students failed: ", fail_count)
```

```python
#clean names list
print("Task 3 - Simple Data Cleaner")
names = ["Alice", "bob", "CHARLIE"]
cleaned_names = []
for name in names:
  cleaned_name = name.strip().lower()
  cleaned_names.append(cleaned_name)
print("old list: ", names)
print("new list: ", cleaned_names)
```

```python
#length of text
print("Task 4 - Message Length Analyser")
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
  length = len(message)
  print("message: ", message, "| length: ", length)
  if length>=10:
    print("Flag message")
  else:
    print("Not a flag messge")
```

```python
#error message detector
print("Task 5 - Error Message Detector")
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for log in logs:
  if log == "ERROR":
    error_count += 1
print(" Total logs: ", len(logs))
print(" Total errors: ", error_count)
```
