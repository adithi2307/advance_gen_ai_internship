```python
#1. Smart Parking Lot Management System
print("1. Smart Parking Lot Management System")
def manage_parking(capacity, vehicle_logs):

    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    if parked > capacity:
        status = "Full - Over Capacity"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak)
    print("Parking Status:", status)


# Dataset
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(parking_capacity, vehicle_logs)
```

```python
#2. Online Food Delivery Time Estimator
print("2. Online Food Delivery Time Estimator")
def estimate_delivery_time(distance, traffic, weather):

    base_time = distance * 5  # 5 minutes per km

    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10

    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


# Dataset
distance_km = 8
traffic_level = "High"
weather_condition = "Rainy"

estimate_delivery_time(distance_km, traffic_level, weather_condition)
```

```python
#3. Movie Theatre Seat Occupancy Analyzer
print("3. Movie Theatre Seat Occupancy Analyzer")
def analyze_occupancy(total_seats, booked_seats):

    booked_count = 0

    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", str(int(occupancy)) + "%")
    print("Show Status:", status)


# Dataset
total_seats = 200
booked_seats = [1] * 150  # 150 booked seats

analyze_occupancy(total_seats, booked_seats)
```

```python
#4. Cloud Server Load Classification System
print("4. Cloud Server Load Classification System")
def classify_server_load(cpu_readings):

    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", str(int(average)) + "%")
    print("Server Status:", status)


# Dataset
cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

```python
#5. Smart Classroom Resource Usage Monitor
print("5. Smart Classroom Resource Usage Monitor")
def monitor_resource_usage(resources):

    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", alert)


# Dataset
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(resource_usage)
```

```python
#6. Online Event Registration Capacity Controller
print("6. Online Event Registration Capacity Controller")
def manage_event_registration(capacity, registrations):

    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Dataset
event_capacity = 100
total_registrations = 105

manage_event_registration(event_capacity, total_registrations)
```
