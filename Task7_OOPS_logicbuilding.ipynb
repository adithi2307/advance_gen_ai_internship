```python
print("1. Smart Light Controller Program")
# This program controls a smart light

class SmartLight:

    # Constructor to store light name
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default status

    # Method to turn light ON
    def turn_on(self):
        self.status = "ON"
        print(self.light_name, "is", self.status)

    # Method to turn light OFF
    def turn_off(self):
        self.status = "OFF"
        print(self.light_name, "is", self.status)


# Creating object
light1 = SmartLight("Bedroom Light")

# Performing action
light1.turn_on()
```

```python
print("2. Employee ID Card Program")
# This program stores employee details

class Employee:

    # Constructor to store employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Creating object
emp1 = Employee("Rahul", "EMP102", "IT")

# Displaying ID card
emp1.display_id_card()
```

```python
print("3. Mobile Contact Record Program")
# This program stores contact details

class Contact:

    # Constructor to store contact name and phone number
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Creating object
contact1 = Contact("Anita", "9876543210")

# Displaying contact details
contact1.display_contact()
```

```python
print("4. Product Price Tag Program")
# This program stores product details

class Product:

    # Constructor to store product name and price
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    # Method to display price tag
    def display_price(self):
        print("Product:", self.product_name)
        print("Price: ₹", self.price)


# Creating object
product1 = Product("Headphones", 2499)

# Displaying price
product1.display_price()
```

```python
print("5. Movie Rating Program")
# This program stores movie details

class Movie:

    # Constructor to store movie name and rating
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    # Method to display movie details
    def display_movie(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")


# Creating object
movie1 = Movie("Inception", 4.8)

# Displaying movie details
movie1.display_movie()
```

```python
print("6. Delivery Address Program")
# This program stores delivery details

class Delivery:

    # Constructor to store customer name and address
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to display delivery details
    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Creating object
delivery1 = Delivery("Suman", "Hyderabad")

# Displaying delivery details
delivery1.display_delivery()
```
